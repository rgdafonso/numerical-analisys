### Método da bisseção


```
import math
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
```

```


def bissecao(f, a, b, epsilon):
  raizes_bissecao = []
  nr_iteracoes = 0
  if((f(a)*f(b)) <0):   
      x = (a+b)/2
      raizes_bissecao.append(x)
      while(abs(f(x)) > epsilon):
        nr_iteracoes += 1
        if ((f(a) * f(x)) <0):
          b = x
        else:
          a = x
        raizes_bissecao.append(x)
        x = (a+b)/2
      print("A raiz do intervalo dado é ", x)
      print("Número de iterações ", nr_iteracoes)
  else:
    print("Não hã raízes no intervalor dado");
    
  inicio = -1.0
  final = 6.0
  numero_pontos = 100

  curve = np.linspace(inicio,final,numero_pontos+1)

  values = [f(value) for value in curve]

  x_aproximado = [f(value) for value in raizes_bissecao]

  plt.plot(curve, values, linestyle='-')

  plt.plot(raizes_bissecao, x_aproximado, 'x', marker='.', label='' + str(len(raizes_bissecao)) + ' iterações' )

  i=0
  for x,y in zip(raizes_bissecao, x_aproximado):
    i = i+1
    plt.text(x,y,str(i))

  leg = plt.legend();

  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.title('Método da Bisseção')

  plt.axhline(0, color = "black")

  plt.grid()

  plt.show()  
    

      
```

### Método da falsa-posição

### Método da Secante

```
def secante(f, a, b, epsilon):
  nr_iteracoes = 0
  raizes_secante =[]
  if(f(a)*f(b) < 0):
    if(abs(f(a)) < epsilon):
      x = a
      print("A raiz do intervalo é ", x)
      return
  
    if(abs(f(b)) < epsilon and abs(b - a) < epsilon):
      x =b
      print("A raiz do intervalo é ", x)
      return
  
    x2 = b - ((f(b)/(f(b)-f(a))) * (b-a))
    while(abs(f(x2)) > epsilon and abs(b - a) > epsilon):
      nr_iteracoes += 1
      raizes_secante.append(x2)
      x2 = b - ((f(b)/(f(b)-f(a))) * (b-a))
      a = b
      b =x2
      #print("A raiz do intervalo é ", x2)
    x = x2
    print("A raiz do intervalo é ", x)
    print("Número de iterações ", nr_iteracoes)
    
  else:
    print("Não há raizes no intervalo dado.")
    
    
  inicio = -1.0
  final = 6.0
  numero_pontos = 100

  curve = np.linspace(inicio,final,numero_pontos+1)

  values = [f(value) for value in curve]

  x_aproximado = [f(value) for value in raizes_secante]

  plt.plot(curve, values, linestyle='-')

  plt.plot(raizes_secante, x_aproximado, 'x', marker='.', label='' + str(len(raizes_secante)) + ' iterações' )

  i=0
  for x,y in zip(raizes_secante, x_aproximado):
    i = i+1
    plt.text(x,y,str(i))

  leg = plt.legend();

  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.title('Método da Secante')

  plt.axhline(0, color = "black")

  plt.grid()

  plt.show()
    
```

### Método da Tangente

```
def tangente(f, f_d, a, b, epsilon):
  nr_iteracoes = 0
  raizes_tangente = []
  if(f(a)*f(b) < 0):
    if(abs(f(a)) < epsilon):
      x = a
      print("A raiz do intervalo é ", x)
      return
    x1 = a - (f(a) / (f_d(a)))
    while(abs(f(x1)) > epsilon and abs(x1 - a) > epsilon):
      nr_iteracoes +=1
      a = x1
      raizes_tangente.append(x1)
      x1 = a - (f(a) / (f_d(a)))
    print("A raiz do intervalo é ", x1)
    print("Número de iterações ", nr_iteracoes)    
  else:
    print("Não há raizes no intervalo dado.")
    
    
  inicio = -1.0
  final = 6.0
  numero_pontos = 100

  curve = np.linspace(inicio,final,numero_pontos+1)

  values = [f(value) for value in curve]

  x_aproximado = [f(value) for value in raizes_tangente]

  plt.plot(curve, values, linestyle='-')

  plt.plot(raizes_tangente, x_aproximado, 'x', marker='.', label='' + str(len(raizes_tangente)) + ' iterações' )

  i=0
  for x,y in zip(raizes_tangente, x_aproximado):
    i = i+1
    plt.text(x,y,str(i))

  leg = plt.legend();

  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.title('Método da Secante')

  plt.axhline(0, color = "black")

  plt.grid()

  plt.show()
  
```
